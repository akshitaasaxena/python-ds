```python
#pieplot
subjects=['Maths','physics','cs','english']
scores=[30,40,20,10]
plt.pie(scores,labels=subjects,autopct='%1.1f%%',shadow=True,startangle=90,explode=[0,0.1,0,0])
#autopct: portion hai, uska kitne percent hai, voh dikhao
#1.1:point ke baad ek decimal: 3 krdoge to 3 point ayenge no ke baad
#startangle: piechart kis angle se start hoga. 0 krdenge to horizontal line se start
plt.tight_layout()
```

```python
#box plot
batch1=np.random.randint(10,70,100)
batch2=np.random.randint(10,80,100)

plt.boxplot((batch1,batch2),tick_labels=['batch1','batch2']) #boxplot ek hi parameter lega
plt.tight_layout()
plt.title('plot')
plt.xlabel('x-axis')#labeling of x and y axis
plt.ylabel('y-axis')
```

```python
df=pd.read_csv("student_data.csv")
```

```python
sns.lineplot(data=df,x='study_hours',y='exam_score')
plt.tight_layout()
#apne aap sb aagya label etc
#line ke paas ka fog: datapoints is line me lie kr rhe hain
plt.show()
```

```python
def grade(score):
  if score>=90:
    return "A"
  elif score >= 80:
    return "B"
  elif score >= 70:
    return "C"
  elif score >= 60:
    return "D"
  else:
    return "F"
df['grade'] = df['exam_score'].apply(grade)
```

```python
sns.barplot(data=df,x='grade',y='exam_score',estimator='max')#estimator: box plot pe jo apply krne ki koshish krre ho voh hai kya
plt.title("performance")
```
