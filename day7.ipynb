```python
#leetcode 1757
import pandas as pd

def find_products(products: pd.DataFrame) -> pd.DataFrame:
    return products.loc[
    (products['low_fats'] == 'Y') & (products['recyclable'] == 'Y'),
    ['product_id']]
```

```python
#leetcode 584
import pandas as pd

def find_customer_referee(customer: pd.DataFrame) -> pd.DataFrame:
    return customer.loc[customer['referee_id'].isna() | (customer['referee_id']!=2),['name']]
```

```python
#leetcode 595
import pandas as pd

def big_countries(world: pd.DataFrame) -> pd.DataFrame:
    return world.loc[(world['area']>=3000000)|(world['population']>=25000000),['name','population','area']]
```

# GROUP-BY

```python
import pandas as pd
df=pd.read_csv('student_data.csv')
df.groupby('placement_status')['placement_status'].agg('count')
```

syntax: df.groupby('kis column by group by')[kis column pe operation perform krna hai'].agg(function)

zaruri nahi hai ki kis pe lagana hai aur jispe operation erformm krna ho vo same ho

```python
df.groupby('placement_status')['exam_score'].agg('mean')
```

```python
df.groupby('placement_status')['sleep_hours'].agg('mean')
```

```python
df.groupby('study_hours')['exam_score'].agg('mean')
```

# APPLY FUNCTION

```python
def grades(c):
  if c>=90:
    return "a"
  elif c>=80:
    return "b"
  elif c>=70:
    return "c"
  else:
    return "fail"
print(grades(90))
df['grade']=df['exam_score'].apply(grades) # 'grade'= ye column ka name hai jo add hojayega, grades function ka name hai jo apply hoga
df
```
