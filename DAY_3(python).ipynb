````python
y="  hi "
h="  helloo  "
i=" from herE"

p= y.strip()+','+h.strip()+ ' '+ i.strip()
print(p)
l=p.capitalize()
print(l)
#REPLACE
g=l.replace("here","there")
print(g)

#find-finds text, returns index, if doesnot find- return (-19)
a=g.find("there")
print(a)
````

````python
#check text- startswith(), endwith()
u="coming here from delhi"
y=u.startswith("here")
print(y)
h=u.endswith("delhi")
print(h)
````

````python
#valid palindrome-leetcode
s = "race a car"
t=s.lower()
print(t)
k=""
for i in t:

  if i.isalnum():
    k=k+i
print(k)
if (k==k[::-1]):
  print("true")
else:
  print("false")
````
