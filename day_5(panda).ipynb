```python
#ques-2-numpy
stock=np.random.randint(20,101,size=(5,5))
demand=np.random.randint(1,101,size=(5,5))
remstock=stock-demand
print(remstock)
totalshortage=abs(np.sum(np.where(remstock<0,remstock,0),axis=1)) #abs for converting negative values to positive ones.
print(totalshortage)
replace=np.where(remstock<0,0,remstock)
print(replace)
zerostock=np.sum(np.where(replace==0,1,0),axis=1)
print(zerostock)
sort=np.sort(zerostock)
```
