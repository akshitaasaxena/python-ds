```python
#constructor
#init-constructor- apne aap call hota hai jab class banate hain
class student:
  def __init__(self,name,age):
    self.name=name                 # self.name: is class ke andar jo name hai uski baat hori hai
    self.age=age # ek variable hai ek value
  def display(self): #function
    print("name",self.name)
    print("age",self.age)

a=student("akshita",20) # a is the object
print(a.name)
a.display()
# ecapsulation : can be achieved, using classes, in python and js can be done using functions as well.
#abstraction : can be achieved using libraries aur header files.
```

passing the property from one class to another class is inheritance

```python
class Person:
    def __init__(self, h, w, age):
        self.height = h
        self.weight = w
        self.age = age


class Rahul(Person):
    def __init__(self, h, w, age, g, name, b):
        super().__init__(h, w, age)   # Person class constructor
        self.gender = g
        self.name = name
        self.bank_balance = b


rh = Rahul(170, 65, 20, "M", "Rahul", 50000)

print(rh.gender)
print(rh.name)
print(rh.bank_balance)
print(rh.height)
```

```python
class complex :
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def __add__(self,other):
    a = self.x + other.x
    b = self.y + other.y
    return complex (a,b)

  def display(self):
    print(self.x, '+',self.y,'i')
a = complex(2,3)
b = complex(4,5)
c = a + b
c.display()
```
